```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
```

```python
city_data_to_load = "/Users/georgesehremelis/Desktop/Data_Science_Bootcamp/pyber.py/05_matplotlib_homework_Instructions_Pyber_data_city_data.csv"
ride_data_to_load = "/Users/georgesehremelis/Desktop/Data_Science_Bootcamp/pyber.py/05_matplotlib_homework_Instructions_Pyber_data_ride_data.csv"
```

```python
city_file = pd.read_csv(city_data_to_load)
ride_file = pd.read_csv(ride_data_to_load)
ride_file.head()
```

```python
merge_files = pd.merge(city_file, ride_file, on="city")


merge_files.head()
urban_plots = merge_files[merge_files["type"] == 'Urban']
suburban_plots = merge_files[merge_files["type"] == 'Suburban']
rural_plots = merge_files[merge_files["type"] == 'Rural']

city_groupU = urban_plots.groupby('city')
city_groupS = suburban_plots.groupby('city')
city_groupR = rural_plots.groupby('city')

ux = city_groupU.count()["ride_id"]
us = city_groupS.count()["ride_id"]
ur = city_groupR.count()["ride_id"]

dcountU = city_groupU.mean()["driver_count"]
dcountS = city_groupS.mean()["driver_count"]
dcountR = city_groupR.mean()["driver_count"]

ave_fareU = city_groupU.mean()["fare"]
ave_fareS = city_groupS.mean()["fare"]
ave_fareR = city_groupR.mean()["fare"]

plt.scatter(ux, ave_fareU, marker="o", facecolors="red", edgecolors="black",
              s= dcountU*10, alpha=0.75)

plt.scatter(us, ave_fareS, marker="o", facecolors="orange", edgecolors="black",
              s=dcountS*10, alpha=0.75)

plt.scatter(ur, ave_fareR, marker="o", facecolors="blue", edgecolors="black",
             s=dcountR*10, alpha=0.75)
```

```python
urban_plots = merge_files[merge_files["type"] == 'Urban']
suburban_plots = merge_files[merge_files["type"] == 'Suburban']
rural_plots = merge_files[merge_files["type"] == 'Rural']

u = urban_plots.sum()["fare"]
s = suburban_plots.sum()["fare"]
r = rural_plots.sum()["fare"]

total_sum = u + s + r

percent_urban = u/total_sum
percent_suburban = s/total_sum
percent_rural = r/total_sum

types = [u, s, r]
labels = ["Urban", "Suburban", "Rural"]
colors = ["red", "orange", "blue"]

plt.pie(types, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.show()
```

```python
uU = urban_plots.count()["city"]
uS = suburban_plots.count()["city"]
uR = rural_plots.count()["city"]

ride_types = [uU, uS, uR]
plt.pie(ride_types, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.show()
```
